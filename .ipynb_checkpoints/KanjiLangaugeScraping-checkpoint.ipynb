```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

url1 = 'http://www.kanjidamage.com/kanji'

reqKD = requests.get(url1)   #request for scraping the url which returns html status code 

url2 = 'https://www.nihongo-pro.com/kanji-pal/list/jlpt'

reqNP = requests.get(url2)   

soup = bs(reqKD.content, 'html.parser')   #parsing the html content 

elements = soup.select("table tr")   #select a particular content which want to scrape
numbers = []
for i in elements:
    numbers.append(int(i.td.text))  # here the contents are in text form so appending to the list 

elements = soup.select("table tr a")  
symbols = []
for i in elements:
    symbols.append(i.text)

kanjis = {}
for i in range(len(numbers)):
    kanjis[numbers[i]] = symbols[i]

soup = bs(reqNP.content, 'html.parser')
elements = soup.select(".outputListContents span")

#N3 has 617 Kanjis according to NihongoPro
reqdKanjis = 617
n3kanjis = []

for i in range(reqdKanjis):
    n3kanjis.append(elements[i].text)
    
df = pd.DataFrame()

no = []
kanji = []

for i,j in kanjis.items():
    
    if j in n3kanjis:
        no.append(i)
        kanji.append(j)
    
df["Number"] = no
df["Kanji"] = kanji
df.to_csv("kanjiDamage.csv",index=False)
```
